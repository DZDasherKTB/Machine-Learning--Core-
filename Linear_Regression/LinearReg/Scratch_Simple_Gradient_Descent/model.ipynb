```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples = 4,n_features=1,n_informative=1,noise = 80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
#applying gradient descent assuming slope is constant m = 78.35 
#assuming intial intercept b = 100
y_pred = ((78.35*x)+ 100).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='green',label = 'b = 100')
plt.legend()
plt.show()
```

```python
m = 78.5
b = 100

loss_slope = -2* np.sum(y -m*x.ravel() -b)
loss_slope
```

```python
lr = 0.05
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred = ((78.35*x)+ b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x, y_pred, color='green', label=f'b = {b:.2f}')
plt.legend()
plt.show()
```

```python
epoch = 20
lr = 0.05
b_vals = []

for i in range(epoch):
    loss_slope = -2 * np.sum(y - m * x.ravel() - b)
    step_size = loss_slope * lr
    b = b - step_size
    y_pred = (m * x + b).reshape(4)

    plt.scatter(x, y)
    plt.plot(x, reg.predict(x), color='red', label='OLS')
    plt.plot(x, y_pred, color='green', label=f'b = {b:.2f}')
    plt.title(f'Epoch {i+1}')
    plt.legend()
    plt.grid(True)
    plt.pause(0.5)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
df = pd.DataFrame(data=dataset['data'], columns = dataset['feature_names'])
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['diagnosis'] = ['M' if i==0 else 'B' for i in dataset['target']]
```

```python
df.sample(5)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
x_train
```

```python
x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = knn.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
print(x_test.shape)
print(y_test.shape)
print(y_pred.shape)
```

```python
from sklearn.model_selection import GridSearchCV
param = {'n_neighbors': list(range(1,31))}
grid_search = GridSearchCV(
  estimator=knn,param_grid=param,cv = 5, scoring= 'accuracy'
)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_knn = grid_search.best_estimator_
y_pred = best_knn.predict(x_test)
```

```python
print("Best number of neighbors:", grid_search.best_params_['n_neighbors'])
print("Best cross-validation accuracy:", grid_search.best_score_)
```

```python
accuracy_score(y_test,y_pred)
```
